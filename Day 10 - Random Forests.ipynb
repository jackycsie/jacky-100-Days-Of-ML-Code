```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
dataset = pd.read_csv('./dataset/Social_Network_Ads.csv')
dataset
```

```python
X_numerical = dataset.iloc[:, 2:4]
X_categoral = dataset.iloc[:, 1]
y = dataset.iloc[:, 4].values
```

# 將性別做分類

```python
label_mapping = {'Male':1,'Female':0}
X_categoral_label = X_categoral.map(label_mapping)
X_categoral_label
```

# 將性別合併和數執行資料合併

```python
X = pd.concat([X_numerical, X_categoral_label], axis=1)
X = X.values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

# 將資料做標準化

```python
sc = StandardScaler()
X_train = sc.fit_transform(pd.DataFrame(X_train[:, :2]))
X_test = sc.fit_transform(pd.DataFrame(X_test[:, :2]))
```

# 訓練模型，50棵數，若feature小於等於2就不在切割，或者葉子中的資料為1

```python
classifier = RandomForestClassifier(n_estimators = 9, criterion = 'entropy', min_samples_split = 2, min_samples_leaf = 1, random_state = 25)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```
